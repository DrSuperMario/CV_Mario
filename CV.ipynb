```python
import pandas as pd
import matplotlib.pyplot as plt
#matplotlib image module
import matplotlib.image as imgRead
```

```python
#reading in the image
img = imgRead.imread('ink.png')
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5)) #add two subplots with figsize
explode = (0, 0.1, 0, 0, 0) #ecplode certan part
num = [20,20,20,20,20]
lables = ['hummm','kumm','summm','trumm','solll']
ax[0].imshow(img)
ax[1].pie(num, labels=lables, explode=explode, shadow=True, autopct='%1.1f%%');
```
